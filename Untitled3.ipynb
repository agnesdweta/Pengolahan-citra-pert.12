```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("daunhijau.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
blur = cv2.GaussianBlur(gray, (5,5), 0)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Citra Asli")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Grayscale + Gaussian Blur")
plt.imshow(blur,cmap='gray')
plt.axis("off")

plt.show()
```

```python
sobelx = cv2.Sobel(blur,cv2.CV_64F,1,0,ksize=3)
sobely = cv2.Sobel(blur,cv2.CV_64F,0,1,ksize=3)

sobel = cv2.magnitude(sobelx,sobely)
```

```python
canny = cv2.Canny(blur, 100, 200)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Sobel")
plt.imshow(sobel,cmap='gray')

plt.subplot(1,2,2)
plt.title("Canny")
plt.imshow(canny, cmap='gray')

plt.show()
```

```python
_, thresh = cv2.threshold( blur,0,255, cv2.THRESH_BINARY+cv2.THRESH_OTSU )
```

```python
import numpy as np
kernel = np.ones((5,5),np.uint8)
```

```python
erosi = cv2.erode(thresh,kernel,iterations=1)
```

```python
dilasi = cv2.dilate(thresh,kernel,iterations=1)
```

```python
opening = cv2.morphologyEx( thresh, cv2.MORPH_OPEN, kernel )
```

```python
closing = cv2.morphologyEx( thresh, cv2.MORPH_CLOSE, kernel )
```

```python
titles = ["Threshold", "Erosi", "Dilasi", "Opening", "Closing"]
images = [thresh, erosi, dilasi, opening, closing]

plt.figure(figsize=(15,5))

for i in range(5):
    plt.subplot(1,5,i+1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap='gray')
    plt.axis("off")

plt.show()
```

```python
hist = cv2.calcHist([gray], [0], None, [256], [0,256])

plt.plot(hist)
plt.title("Histogram Intensitas")
plt.xlabel("Pixel")
plt.ylabel("Frekuensi")
plt.show()
```

```python
from skimage.feature import graycomatrix
from skimage.feature import graycoprops
glcm = graycomatrix( gray, [1], [0], 256, symmetric=True, normed=True )

contrast = graycoprops(glcm, 'contrast')[0,0]
dissimilarity = graycoprops(glcm, 'dissimilarity')[0,0]
homogeneity = graycoprops(glcm, 'homogeneity')[0,0]
```

```python
contours,_ = cv2.findContours( closing, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE )

cnt = max(contours,key=cv2.contourArea)

area = cv2.contourArea(cnt)
perimeter = cv2.arcLength(cnt, True)

circularity = (4*np.pi*area) / (perimeter**2)
print(area)
print(perimeter)
print(circularity)
```

```python
result = img_rgb.copy()
cv2.drawContours( result, [cnt], -1, (255,0,0), 3 )

plt.imshow(result)
plt.title("Kontur Objek Utama")
plt.axis("off")
plt.show()
```
